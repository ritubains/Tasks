```python
data_copy=data.copy()

data_copy.Pclass.replace([1, 3], [3, 1], inplace=True)
```

```python
sns.heatmap(data_copy.corr(),annot=True)
```

```python
family_size=data_copy['Siblings/Spouses Aboard']+data_copy['Parents/Children Aboard']
```

```python
family_size
```

```python
data_copy = pd.concat([ data_copy,family_size],axis=1)

```

```python
data_copy.rename(columns={0: 'Family size'}, inplace=True)
data_copy.head(5)
```

```python
X = data_copy.drop(['Survived'],axis = 1)
y = data_copy['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
 from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
