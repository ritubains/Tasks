```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2=fig.add_axes([0.0,0.5,0.4,0.3])
axes2.plot(x,y)
axes2.set_title('short Plot')
```
